## Part 2: Your First AI Prompt

Use an AI assistant (Gemini in Colab, ChatGPT, or GitHub Copilot) to create a program that:
- Asks the user for their name
- Asks for their monthly income  
- Prints a personalized welcome message

### 📝 Document Your AI Prompt Below:

```python
# TASK: Write your AI prompt as a comment below
"""
My AI Prompt:
I am a beginner learning Python programming.
Please create a simple Python program that:

1. Asks the user to enter their name
2. Asks the user to enter their monthly income
3. Displays a personalised welcome message using their name
4. Prints the income entered by the user

Use simple beginner-friendly Python code with input() and print().
"""

# Paste the AI-generated code below:
name = input("What is your name? ")
monthly_income_str = input("What is your monthly income? ")

# It's good practice to convert income to a number if you plan to do calculations
try:
    monthly_income = float(monthly_income_str)
    print(f"Hello, {name}! Welcome to your personal finance assistant.")
    print(f"Your monthly income is: ${monthly_income:.2f}")
except ValueError:
    print(f"Hello, {name}! Welcome to your personal finance assistant.")
    print(f"Your monthly income is: {monthly_income_str} (Please enter a number for calculations later)")
```
